```python
# --- FILE check_nghi_dinh.py ---
from src.config import embeddings, PERSIST_DIRECTORY
from langchain_chroma import Chroma

vector_db = Chroma(persist_directory=PERSIST_DIRECTORY, embedding_function=embeddings)

# Lấy toàn bộ dữ liệu ra để check
collection_data = vector_db.get()

print("--- DANH SÁCH CÁC ĐIỀU LẠC VÀO NGHỊ ĐỊNH XỬ PHẠT ---")
count = 0
for i in range(len(collection_data['ids'])):
    meta = collection_data['metadatas'][i]
    if meta.get('source') == "Nghị định xử phạt":
        print(f"-> {meta.get('chuong')} | {meta.get('dieu')}")
        count += 1

print(f"\n Total: Tổng số điều thực tế của Nghị định xử phạt trong DB là: {count}")
```

```python
import os
from dotenv import load_dotenv
from neo4j import GraphDatabase

load_dotenv()

uri = os.getenv("NEO4J_URI")
user = os.getenv("NEO4J_USERNAME")
pwd = os.getenv("NEO4J_PASSWORD")

print(f"🔄 Đang thử kết nối tới: {uri}...")
try:
    with GraphDatabase.driver(uri, auth=(user, pwd)) as driver:
        driver.verify_connectivity()
    print("🎯 Kết nối thành công! Đã thông mạng tới Neo4j AuraDB.")
except Exception as e:
    print(f"❌ Thất bại. Lỗi chi tiết: {e}")
```

```python
from src.retriever import HybridVectorGraphRetriever

def test():
    retriever = HybridVectorGraphRetriever()
    
    # Thử đặt một câu hỏi liên quan đến nội dung trong Luật Đường bộ bạn vừa nạp
    query = "Quy định về hành vi bị nghiêm cấm trên đường bộ"
    
    results = retriever.retrieve(query, top_k=2)
    
    print("\n" + "="*50 + "\n🎯 KẾT QUẢ TRUY VẤN LAI (HYBRID CONTEXT):\n" + "="*50)
    for idx, res in enumerate(results):
        print(f"\n[Mảnh gốc {idx+1}] Nguồn: {res['source']} | {res['chuong']}")
        print(f"📌 Vị trí: {res['dieu']}")
        print(f"📄 Nội dung ngữ cảnh (Đã được Neo4j khôi phục trọn vẹn):\n{res['content']}")
        print("-" * 40)

if __name__ == "__main__":
    test()
```
